```python
import numpy as np
import matplotlib.pyplot as plt

function_x = np.linspace(-3, 3, 100)
function_y = 2 * function_x ** 2

fig, ax = plt.subplots(figsize=(10, 5))
# ax.plot(function_x, function_y) # 실제로는 이 그래프를 모르는 상태

x = 2
y = 2 * x ** 2
ax.scatter(x, y, color='red', s=200)
lr = 0.1

for _ in range(5):
    dy_dx = 4 * x
    x = x - lr * dy_dx
    y = 2 * x ** 2
    
    ax.scatter(x, y, color='red', s=100)
```
